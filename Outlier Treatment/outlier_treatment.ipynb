```python
#create the dataframe
boston_df = pd.DataFrame(boston.data)
boston_df.columns = columns
```

```python
import seaborn as sns
sns.boxplot(x=boston_df['DIS'])
#sns.boxplot(x=boston_df['INDUS'])
#sns.boxplot(x=boston_df['CRIM'])
#sns.boxplot(x=boston_df['ZN'])
```

```python
##IQR = Q3 − Q1.
Q1 = boston_df.quantile(0.25)
Q3 = boston_df.quantile(0.75)
```

```python
IQR = Q3 - Q1
print(IQR)
```

```python
boston_df_out = boston_df[~((boston_df < (Q1 - 1.5 * IQR)) |(boston_df > (Q3 + 1.5 * IQR))).any(axis=1)]
boston_df_out.shape
```

```python
sns.boxplot(x=boston_df_out['DIS'])
#sns.boxplot(x=boston_df_out['INDUS'])
#sns.boxplot(x=boston_df_out['CRIM'])
#sns.boxplot(x=boston_df_out['ZN'])
```
